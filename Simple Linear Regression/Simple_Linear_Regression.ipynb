```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_dataset.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Years_Of_Experience'], df['Salary'])
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
X = df[['Years_Of_Experience']] # X = df.iloc[:,0].values
y = df[['Salary']] # y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
class LinearRegression:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self,X_train,y_train):
        num = 0
        den = 0
        x_mean = X_train.iloc[:,0].mean()
        y_mean = y_train.iloc[:,0].mean()

        for i in range(X_train.shape[0]):
            num = num + ((X_train.iloc[i,0] - x_mean) * (y_train.iloc[i,0] - y_mean))
            den = den + ((X_train.iloc[i,0] - x_mean) * (X_train.iloc[i,0] - x_mean))
        
        self.m = num/den
        self.b = y_mean - (self.m * x_mean)
        print("The value of b is: ", self.b)
        print("The value of m is: ", self.m)

    def predict(self,X_test):
         y = self.m * X_test + self.b
         return y

lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test.iloc[0,0])
```

```python
plt.scatter(df['Years_Of_Experience'],df['Salary'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('Years Of Experience')
plt.ylabel('Salary')
plt.show()
```
